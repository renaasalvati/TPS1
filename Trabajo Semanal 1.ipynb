**2. ¿Qué tipo de filtro es?**

Si se hace la relación $\frac{R_2}{R_1}$ unitaria, se va a obtener una misma frecuencia natural para los polos y los ceros. De esta va a haber una imagen especular en el diagrama, lográndose un filtro pasa todo.

**3. Obtenga la función transferencia, pero normalizada. ¿Cuál sería en este caso la norma de frecuencia y qué interpretación circuital podría tener?**

Para este caso se puede normalizar a la frecuencia natural de los polos, en donde si la relación $\frac{R_2}{R_1}$ es unitaria esta misma frecuencia natural sería para los ceros.

\begin{equation}
\omega_0 = \frac{1}{C_1 R_3} \qquad  
S' = \frac{S}{\omega_0} \rightarrow S = S' \omega_0
\end{equation}

\begin{equation}
H(s) = \frac{S'\omega_0-\frac{R_2}{R_1}\omega_0}{S'\omega_0+\omega_0} = \frac{S'-\frac{R_2}{R_1}}{S'+1}
\end{equation}

Si $\frac{R_2}{R_1} = 1$:

\begin{equation}
H(s) = \frac{S'-1}{S'+1}
\end{equation}
